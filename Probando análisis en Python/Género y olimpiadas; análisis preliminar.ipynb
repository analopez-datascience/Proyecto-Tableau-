```python
import warnings
warnings.filterwarnings("ignore")

import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./athlete_events.csv')
df.head(10)
```

```python

# Filtrar filas donde hay medallas
df_medallas = df[df['Medal'].notna()]

# Agrupar por año y género, y contar el número de medallas
medallas_por_genero = df_medallas.groupby(['Year', 'Sex'])['Medal'].count().unstack(fill_value=0)

print("Medallas por Año y Género:")
print(medallas_por_genero.head(40))
```

```python
# Agrupar por año y género, y contar el número de participantes
participacion_por_genero = df.groupby(['Year', 'Sex'])['ID'].count().unstack(fill_value=0)

print("Participación por Año y Género:")
print(participacion_por_genero.head(40))
```

```python
# Agrupar por deporte y género, y contar el número de participantes
deportes_por_genero = df.groupby(['Sport', 'Sex'])['ID'].count().unstack(fill_value=0).fillna(0)

print("Participación por Deporte y Género:")
print(deportes_por_genero.head(50))
```

```python
# Filtrar filas donde hay medallas y agrupar por género, calculando la edad promedio
edad_promedio_medallistas = df_medallas.groupby('Sex')['Age'].mean()

print("Edad Promedio de Medallistas por Género:")
print(edad_promedio_medallistas)
```

```python
# Agrupar por género y calcular estadísticas descriptivas de altura y peso
altura_peso_por_genero = df.groupby('Sex')[['Height', 'Weight']].describe()

print("Estadísticas de Altura y Peso por Género:")
print(altura_peso_por_genero)
```

```python


# Contar el número de participantes por género
participantes_por_genero = df.groupby('Sex')['ID'].nunique()

print("Cantidad de Participantes por Género:")
print(participantes_por_genero)
```

```python

# Filtrar filas donde hay medallas
df_medallas = df[df['Medal'].notna()]

# Agrupar por temporada, género, y contar el número de medallas
medallas_por_temporada_genero = df_medallas.groupby(['Season', 'Sex'])['Medal'].count().unstack(fill_value=0)

print("Medallas por Temporada y Género:")
print(medallas_por_temporada_genero)
```

```python
# Agrupar por temporada, género, y contar el número de medallas
medallas_por_temporada_genero = df_medallas.groupby(['Season', 'Sex'])['Medal'].count().unstack(fill_value=0)

medallas_por_temporada_genero
```
