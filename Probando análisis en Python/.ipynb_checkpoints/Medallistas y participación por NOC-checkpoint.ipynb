```python
import warnings
warnings.filterwarnings("ignore")

import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./athlete_events.csv')
df.head(10)
```

```python

# Contar el número de participantes únicos por NOC
participantes_por_noc = df.groupby('NOC')['ID'].nunique().reset_index(name='Participantes')

print("Cantidad de Participantes por NOC:")
print(participantes_por_noc.head())
```

```python

# Contar el número de participantes únicos por NOC
participantes_por_noc = df.groupby('NOC')['ID'].nunique().reset_index(name='Participantes')

# Ordenar por el número de participantes en orden descendente y seleccionar los primeros cinco
ranking_top5_noc = participantes_por_noc.sort_values(by='Participantes', ascending=False).head(5)

print("Ranking de los cinco NOC con más participantes:")
print(ranking_top5_noc)
```

```python
# Ordenar por el número de participantes en orden descendente y seleccionar los primeros cinco
ranking_top5_noc = participantes_por_noc.sort_values(by='Participantes', ascending=False).head(5)

ranking_top5_noc
```

```python


# 1. Cantidad de Participantes por NOC
participantes_por_noc = df.groupby('NOC')['ID'].nunique().reset_index(name='Participantes')
print("Cantidad de Participantes por NOC:")
print(participantes_por_noc.head(30), "\n")

# 2. Cantidad de Medallas por NOC
df_medallas = df[df['Medal'].notna()]
medallas_por_noc = df_medallas.groupby('NOC')['Medal'].count().reset_index(name='Medallas')
print("Cantidad de Medallas por NOC:")
print(medallas_por_noc.head(30), "\n")

# 3. Distribución de Participación por Género en cada NOC
participacion_por_noc_genero = df.groupby(['NOC', 'Sex'])['ID'].nunique().unstack(fill_value=0)
print("Participación por NOC y Género:")
print(participacion_por_noc_genero.head(30), "\n")

# 4. Cantidad de Medallas por NOC y Género
medallas_por_noc_genero = df_medallas.groupby(['NOC', 'Sex'])['Medal'].count().unstack(fill_value=0)
print("Cantidad de Medallas por NOC y Género:")
print(medallas_por_noc_genero.head(30), "\n")
```

```python


# Contar el número de participantes por NOC y género
participacion_por_noc_genero = df.groupby(['NOC', 'Sex'])['ID'].nunique().unstack(fill_value=0)

# Crear un DataFrame con el ranking de participantes por género
ranking_participantes = participacion_por_noc_genero.sort_values(by=['F', 'M'], ascending=False)

print("Ranking de Países con Más Participantes por NOC y Género:")
print(ranking_participantes.head(30))
```

```python
# Filtrar filas donde hay medallas
df_medallas = df[df['Medal'].notna()]

# Contar el número de medallistas por NOC y género
medallas_por_noc_genero = df_medallas.groupby(['NOC', 'Sex'])['ID'].nunique().unstack(fill_value=0)

# Crear un DataFrame con el ranking de medallistas por género
ranking_medallistas = medallas_por_noc_genero.sort_values(by=['F', 'M'], ascending=False)

print("Ranking de Países con Más Medallistas por NOC y Género:")
print(ranking_medallistas.head(30))
```
